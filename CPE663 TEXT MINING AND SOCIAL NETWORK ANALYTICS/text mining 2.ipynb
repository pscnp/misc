```python
def remove_html(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()
```

```python
review_text = df['review']
review_text = review_text.apply(lambda x: remove_html(x))
review_text = review_text.apply(lambda x: x.strip())
review_text = review_text.apply(lambda x: x.lower())
review_text = review_text.apply(lambda x: x.translate(str.maketrans(string.punctuation, ' '*len(string.punctuation))))
review_text = review_text.apply(lambda x: word_tokenize(x))
review_text = review_text.apply(lambda x: [i for i in x if i not in stopwords.words("english")])
lemmatizer = WordNetLemmatizer()
#review_text = review_text.apply(lambda x: [lemmatizer.lemmatize(i, pos="v") for i in x])
```

```python
review_text
```

```python
w2v_model = gensim.models.Word2Vec(sentences=review_text)
```

```python
w2v_model.wv.most_similar('catholic')
```

```python
y = df['sentiment']
```

```python
y = y.replace('positive', 1)
y = y.replace('negative', 0)
```

```python
model = Sequential()
model.add(Bidirectional(LSTM(32, return_sequences=True, input_shape=(50000, 100, 1))))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['acc'])
```

```python
model = Sequential()
model.add(Embedding(50000, 100, input_length=100))
model.add(Dropout(0.5))
model.add(Bidirectional(LSTM(64)))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))

model.compile(loss='binary_crossentropy',
              optimizer='adam',
              metrics=['acc'])
```

```python
hist = model.fit(w2v_model.wv.vectors, to_categorical(y),
              batch_size=128,
              epochs=50)
```
