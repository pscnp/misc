```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
```

```python
df = pd.DataFrame({ 'x1': np.append(np.repeat([-1,1], 4), [-1.682,1.682] + [0]*10) ,
                    'x2': np.append(np.repeat([-1,1]*2, 2), [0]*2 + [-1.682,1.682] + [0]*8),
                    'x3': [-1,1]*4 + [0]*4 + [-1.682,1.682] + [0]*6,
                    'y': [66,70,78,60,80,70,100,75,100,80,68,63,65,82,113,100,118,88,100,85]})
```

```python
df['x1^2'] = df['x1']**2
df['x2^2'] = df['x2']**2
df['x3^2'] = df['x3']**2
df['x1x2'] = df['x1']*df['x2']
df['x1x3'] = df['x1']*df['x3']
df['x2x3'] = df['x2']*df['x3']
df['x1x2x3'] = df['x1']*df['x2']*df['x3']
```

```python
df
```

```python
x = df.drop('y', axis=1)
x = sm.add_constant(x)
y = df['y']
```

```python
model = sm.OLS(y, x).fit()
```

```python
model.summary2()
```
