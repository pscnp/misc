```python
import pandas as pd
import statsmodels.api as sm
from scipy.stats import pearsonr
from scipy.stats import spearmanr
import numpy as np
from sklearn.preprocessing import StandardScaler
from statsmodels.formula.api import ols
```

```python
df = pd.read_excel('lect10-lin-reg.xlsx', sheet_name='CO-smoke')
```

```python
df.head()
```

```python
y = df['CO']
x = df.drop('CO', axis=1)
```

```python
xreg = sm.add_constant(x)
```

```python
model = sm.OLS(y, xreg).fit()
```

```python
#check_linreg_residuals(model)
```

```python
model.summary2()
```

```python
x2 = df.drop(['CO','TAR'], axis=1)
x2reg = sm.add_constant(x2)
```

```python
model2 = sm.OLS(y, x2reg).fit()
```

```python
model2.summary2()
```

```python
linear_co = pearsonr(df['TAR'], df['NICOTINE'])[0]
nonlinear_co = spearmanr(df['TAR'], df['NICOTINE'])[0]
print('Pearson’s Correlation =', linear_co)
print('Spearman’s Correlation =', nonlinear_co)
```

```python
bottledf = pd.DataFrame({'carbon': np.repeat([10,12,14], 4),
                        'pressure': np.repeat([25,30]*3, 2),
                        'speed': [200,250]*6,
                        'height': [-4,-1,-1,2,1,3,5,11,9,13,16,21]})
```

```python
bottledf
```

```python
y3 = bottledf['height']
x3 = bottledf.drop('height', axis=1)
```

```python
x3reg = sm.add_constant(x3)
```

```python
x3reg['cp'] = x3reg['carbon']*x3reg['pressure']
x3reg['ps'] = x3reg['pressure']*x3reg['speed']
x3reg['cs'] = x3reg['carbon']*x3reg['speed']
x3reg['cps'] = x3reg['carbon']*x3reg['pressure']*x3reg['speed']
```

```python
x3reg
```

```python
model3 = sm.OLS(y3, x3reg).fit()
```

```python
model3.summary2()
```

```python
scaler_x = StandardScaler()
x3reg_scaled = x3
x3reg_scaled['cp'] = x3reg['cp']
x3reg_scaled['ps'] = x3reg['ps']
x3reg_scaled['cs'] = x3reg['cs']
x3reg_scaled['cps'] = x3reg['cps']
x3reg_scaled = pd.DataFrame(scaler_x.fit_transform(x3reg_scaled), columns=x3reg_scaled.columns)
x3reg_scaled = sm.add_constant(x3reg_scaled)
```

```python
model4 = sm.OLS(y3, x3reg_scaled).fit()
```

```python
model4.summary2()
```

```python
# two way ANOVA test
anova_model = ols('height ~ C(carbon)*C(pressure)', bottledf).fit()
```

```python
anova_model.summary()
```

```python
anova_model = ols('height ~ C(carbon)*C(pressure)', bottledf).fit()
sm.stats.anova_lm(anova_model, typ=3)
```

```python
anova_model = ols('height ~ C(carbon)*C(speed)', bottledf).fit()
sm.stats.anova_lm(anova_model, typ=2)
```

```python
anova_model = ols('height ~ C(pressure)*C(speed)', bottledf).fit()
sm.stats.anova_lm(anova_model, typ=2)
```
